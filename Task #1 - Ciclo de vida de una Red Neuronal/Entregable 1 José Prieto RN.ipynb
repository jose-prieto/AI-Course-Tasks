```python
#Primero correr el código: "pip install tensorflow", ya que, no se encuentra instalado de forma predeterminada
from sklearn.model_selection import train_test_split
from keras.layers import Dense, Activation
from keras.models import Sequential
from sklearn import preprocessing
import pandas as pd
import numpy as np
```

```python
#Procesamos los datos
df = pd.read_csv(r"C:\Users\jap25\OneDrive\Escritorio\Entregable_1_Resultados\diabetes.csv")
```

```python
df
```

```python
#Imprimimos la forma del dataframe
df.shape
```

```python
#Homologamos los nombres de las columnas
df.columns = [x.lower().strip() for x in df.columns]
df
```

```python
#Tratamos los datos
df['glucose'] = df['glucose'].replace(0,None)
df['bloodpressure'] = df['bloodpressure'].replace(0,None)
df['skinthickness'] = df['skinthickness'].replace(0,None)
df['insulin'] = df['insulin'].replace(0,None)
df['bmi'] = df['bmi'].replace(0,None)
```

```python
#Tratamos los datos
df['glucose'] = df['glucose'].fillna(df['glucose'].mean())
df['bloodpressure'] = df['bloodpressure'].fillna(df['bloodpressure'].mean())
df['skinthickness'] = df['skinthickness'].fillna(df['skinthickness'].mean())
df['insulin'] = df['insulin'].fillna(df['insulin'].mean())
df['bmi'] = df['bmi'].fillna(df['bmi'].mean())
```

```python
#Procedemos a estandarizan los datos para que cada uno de los valores sean lo más significativos posible
df_escalado = preprocessing.scale(df) #Primero se procesa el escalado
df_escalado = pd.DataFrame(df_escalado, columns=df.columns) #Luego, se convierte el resultado en dataframe para más fácil manejo
df_escalado["outcome"] = df["outcome"] #La salida se mantiene para no afectar resultados
df = df_escalado #Se ambian los valores del dataframe original al ya tratado
df
```

```python
#Separamos los datos de los resultados
x = df.loc[:, df.columns != "outcome"]
y = df.loc[:, df.columns == "outcome"]
```

```python
x
```

```python
y
```

```python
#Ajustamos la configuración de la red neuronal
modelo =  Sequential()
modelo.add(Dense(10, activation='relu', input_dim=8)) #Primera capa
modelo.add(Dense(20, activation='relu')) #Segunda capa
modelo.add(Dense(10, activation='relu')) #Tercera capa
modelo.add(Dense(1, activation='sigmoid')) #Capa de salida
```

```python
#Compilamos la topología de la red neuronal
modelo.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
history = modelo.fit(x_train, y_train, batch_size=10,epochs=100)
```

```python
scores = modelo.evaluate(x_train, y_train)
scores[1]*100
```

```python
scores = modelo.evaluate(x_test, y_test)
scores[1]*100
```
